```python
import cv2 as open_cv
import numpy as numpy
import matplotlib.pyplot as pyplot

import steganography as steg
import timestamp_appender as ts_appender
```

```python
import steganography
data_to_be_hidden = 'Hello World!'
# image_path='images/onder.jpg'
image_path='images/lenna.png'
print('Data to be hidden: ', data_to_be_hidden)

orginal_image = open_cv.imread(image_path, open_cv.IMREAD_GRAYSCALE)
orginal_image = open_cv.resize(orginal_image, (128, 128))

p_prime_number, data_hidden_image = steg.hide_data_to_image(data_to_be_hidden, image_path)

data_hidden_image_path = ts_appender.append_timestamp_to_file_name(image_path)
print("Data hidden image path: ", data_hidden_image_path)
open_cv.imwrite(data_hidden_image_path, data_hidden_image)

pyplot.subplot(121)
pyplot.imshow(orginal_image, cmap='gray')
pyplot.title('Original Image')

pyplot.subplot(122)
pyplot.imshow(data_hidden_image, cmap='gray')
pyplot.title('Data Hidden Image')


extracted_data =  steg.extract_data_from_image(data_hidden_image_path, p_prime_number)
print('Extracted data: ', extracted_data)
```
